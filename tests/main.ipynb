```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.arange(0,20)
y = np.array([0.01, 0.1, 0.15, 0.20, 0.20, 0.29, 0.33, 0.30, 0.29, 0.35, 0.40, 0.456, 0.444, 0.50, 0.489, 0.40, 0.555, 0.532, 0.589, 0.67])
plt.plot(x, y, c="blue", marker="o", label="train")
plt.plot(x, y * np.random.rand(20), c="red", marker="s", label="test")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("F1 Score")
plt.title("F1 over epoch")
```
